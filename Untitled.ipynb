```python
# Peki ben bu görseli numarik bir şekilde görselleştirebilir miyim?

def pixel_visualize(img):
    fig = plt.figure(figsize=(12,12))

    ax = fig.add_subplot(111)
    ax.imshow(img, cmap='gray') 
    width, height = img.shape
    
    threshold = img.max() / 2.5
    for x in range(width):
        for y in range (height):
            ax.annotate(str(round (img[x][y], 2)), xy=(y,x),
                color='white' if img[x][y]<threshold else 'black')
```

# Veriyi hazırlama

```python
# Encoding burda yaptığım bağımlı değişken tanımlamasını yapmak.
# ' ye karşılık gelecek olan yerde bir tanım geldiğinde binary değerde 1 olarak o durumun gerçekteştiğini ifade ediyor.'
[1,2,3,4,5,6,7,8,9]
[0,1,0,0,0,0,0,0,0]
```

```python
# Görsel üzeinden o görsele karşılık gelen değerlerin değerlendirilmesi işlemini yapmaya çalışıyorum.
y_train[0:5]
```

```python
# İfadeleri dönüştürme işlemini yapıyorum.
y_train = to_categorical(y_train)
y_test = to_categorical(y_test)
```

```python
# Burda verilerimin üzerinden bu encoding işlemini gerçekleştirdim işlem öncesi ile kıyaslama yapabilirsiniz.
y_train[0:5]
```

```python
# Reshaping işlemi.
# Bu işeleme yeniden şekillendirme işlemi olarak ifade edebiliriz.
```

```python
# Önce resimlerin boyut bilgisine erişiyorum.
image_size = x_train.shape[1]
image_size
```

```python
# train ve test veri setindeki verileri genel olarak inceleme dahilinbe alıyorum.
print(f"x_train boyutu: {x_train.shape}")
print(f"x_test boyutu: {x_test.shape}")
```

```python
# Burda yeniden şekillendirme işlemi yaparak elimdeki olan verileri alden geçiriyorum.
x_train = x_train.reshape(x_train.shape[0],28,28,1)
x_test = x_test.reshape(x_test.shape[0],28,28,1)

print(f"x_train boyutu: {x_train.shape}")
print(f"x_test boyutu: {x_test.shape}")
# Burdaki 1 sayısı yeni bir kanal bu görsellerdeki değerleri tutmama yarayacak.
# Bu işelemi 28x28 olan görseldeki toplamda 728 olan piksel sayısını ele alarak inceleyebiliriz.
```

```python
# Standartlaştırma işelemi.
# Veiryi belli bir standart hale getirlemliyiz ki eğitim aşamasında veri tek bir standartı olması gerekmektedir.
# Bunun yapmamaızın sebebi eğitim sürecinin daha kısa ve ve eğitim sürecinde daha doğru bir eğitim olmasını sağlamaktadır.
```

```python
# Bu işlemi ise verileri 255 bölerek yapabiliriz tam 255 (beyaz) 1 olur diğerleri ise 0 yada sıfıra yakın değerler alarak işlem yapmaktadır.
x_train = x_train.astype('float32') / 255

x_test = x_test.astype('float32') / 255
# Float 32 dönüşümü programatik acıdan işlerimizi hızlandıracak bir işlem olmaktadır.
```

# Model kurma

```python
# Bu alan sinir ağlarının dış özelikleri.

model = tf.keras.Sequential([
    Flatten(input_shape = (28,28,1)),  # Buraya giriş katmanı.
    Dense(units=128, activation='relu', name='layer1'), #  Burası gizli katman ve burda 128 farklı özelik için 128 ünit tanımlıyorum. relu aktivasyon fonksyonunu kullanıyorum.
    Dense(units=num_labels, activation='softmax', name='output_layer') # Çıkış katmanı görsel özelik sayısı bu değişken olarak çağırdım. Activasyon fonksyonu sınıflandırma ve çok yapılı olduğu için softmax kullandım.
])

# Şimdi geleleim sinir ağının iç özeliklerine.

model.compile(loss='categorical_crossentropy', # Hata sınıflandırmada catogorik işlem yapmasına yararayack.
             optimizer='adam', # Optimizasyon fonsksyonu tanımlanmasını sağlama.
             metrics=[tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), "accuracy"]) # Hata düzeltmedeki metrik tanımlama.
```

```python
# Model özetini incelemeye çalışıyorum.

model.summary()
```

```python
# Şimdi modeli fit etmeye başlıyorum.

model.fit(x_train, y_train, epochs=10, batch_size=128, validation_data=(x_test, y_test))
# Modelin eğitimi için train veri seti tanımlanması yapılıyor.
# epoch sayısı modelin eğitimde bir birimlik işlem yapma sayısı tanımlanamsı.
# batch sayısı işlem sırasında yapılan grandyel işlemlerde kaç veri kümesi bir işlemde ele alınsın. 
# validasyon datası olarak ise test data gruplarımı tanımlıyorum.
```

# MODEL BAŞARIM DEĞERLENDİRME

```python
history = model.fit(x_train, y_train, epochs=10, batch_size=128, validation_data=(x_test, y_test))
```

```python
import matplotlib.pyplot as plt

# --------Grafik 1 Accuracy-----------
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.plot(history.history['accuracy'], color='b', label='Training Accuracy')
plt.plot(history.history['val_accuracy'], color='r', label='Validation Accuracy')
plt.legend(loc='lower right')
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Accuracy', fontsize=16)
plt.ylim([min(plt.ylim()), 1])
plt.title('Eğitim ve Test Başarım Grafiği', fontsize=16)

# --------Grafik 2 Loss--------------
plt.subplot(1, 2, 2)
plt.plot(history.history['loss'], color='b', label='Training Loss')
plt.plot(history.history['val_loss'], color='r', label='Validation Loss')
plt.legend(loc='upper right')
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Loss', fontsize=16)
plt.ylim([0, max(plt.ylim())])
plt.title('Eğitim ve Test Kayıp Grafiği', fontsize=16)
plt.show()


# Arkadaşlar kusura bakmayan benim modelim veri setini ezberledi ve overfitting oldu bunu farklı ezber bozma yöntemleri ile cözebiliriz.
```

```python
# Aşırı öğrenmenin önüne geçen model

import tensorflow as tf
from tensorflow.keras.layers import Dense, Flatten, Dropout

# Modelin güncellenmiş hali Dropout ile
model_1 = tf.keras.Sequential([
    Flatten(input_shape=(28, 28, 1)),  # Giriş katmanı
    Dense(units=128, activation='relu', name='layer1'),  # Gizli katman
    Dropout(rate=0.5),  # Dropout katmanı, %50 oranında nöronları devre dışı bırakır
    Dense(units=num_labels, activation='softmax', name='output_layer')  # Çıkış katmanı
])

# Modeli derleme
model_1.compile(
    loss='categorical_crossentropy',  # Hata fonksiyonu
    optimizer='adam',  # Optimizasyon fonksiyonu
    metrics=[tf.keras.metrics.Precision(), tf.keras.metrics.Recall(), 'accuracy']  # Metrikler
)
```

```python
history_1 = model_1.fit(x_train, y_train, epochs=10, batch_size=128, validation_data=(x_test, y_test))
```

```python
import matplotlib.pyplot as plt

# --------Grafik 1 Accuracy-----------
plt.figure(figsize=(20, 5))
plt.subplot(1, 2, 1)
plt.plot(history_1.history['accuracy'], color='b', label='Training Accuracy')
plt.plot(history_1.history['val_accuracy'], color='r', label='Validation Accuracy')
plt.legend(loc='lower right')
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Accuracy', fontsize=16)
plt.ylim([min(plt.ylim()), 1])
plt.title('Eğitim ve Test Başarım Grafiği', fontsize=16)

# --------Grafik 2 Loss--------------
plt.subplot(1, 2, 2)
plt.plot(history_1.history['loss'], color='b', label='Training Loss')
plt.plot(history_1.history['val_loss'], color='r', label='Validation Loss')
plt.legend(loc='upper right')
plt.xlabel('Epoch', fontsize=16)
plt.ylabel('Loss', fontsize=16)
plt.ylim([0, max(plt.ylim())])
plt.title('Eğitim ve Test Kayıp Grafiği', fontsize=16)
plt.show()
```

```python
loss, precision, recall, acc = model.evaluate(x_test, y_test, verbose=False)

print("\nTest Accuracy: %.2f%%" % (100.0 * acc))
print("\nTest Loss: %.2f" % loss)
print("\nTest Precision: %.2f%%" % (100.0 * precision))
print("\nTest Recall: %.2f%%" % (100.0 * recall))
```

```python
# Aşırı öğrenmenin önüne geçmeye çalıştığım model.

loss, precision, recall, acc = model_1.evaluate(x_test, y_test, verbose=False)

print("\nTest Accuracy: %.2f%%" % (100.0 * acc))
print("\nTest Loss: %.2f" % loss)
print("\nTest Precision: %.2f%%" % (100.0 * precision))
print("\nTest Recall: %.2f%%" % (100.0 * recall))
```
